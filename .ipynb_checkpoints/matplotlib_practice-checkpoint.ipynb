```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object oriented method number 2
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object oriented method number 3
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend. MATLAB
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot with additional changes. MATLAB
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Add additional changes using the object-oriented approach

# Create the plot with a legend labeled "Boston", line color green, line width 2, and a diamond for a marker.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", linewidth=2, marker= "d", label='Boston')

# Change the marker to a diamond.

# Add a legend for the city of Boston.
ax.legend()

# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Month")

# Add a y-axis limit.
ax.set_ylim(0, 45)

# Add grid lines.
ax.grid()
```

```python
# Create the plot using MATLAB method
plt.bar(x_axis, y_axis)
```

```python
# Create the plot with colors and labels. MATLAB method.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot: horizontal bar chart with MATLAB method.
plt.barh(x_axis, y_axis)
```

```python
# Create the plot with data on opposite axes.
plt.barh(y_axis, x_axis)
```

```python
# Create the plot and invert the y-axis.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot with legend and magenta bar color.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

# Add the legend
plt.legend()

# Add title and axis labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Invert the y-axis.
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt() object-oriented method.
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object-oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object-oriented method and put data on opposite axes.
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create the plot with ax.plt() object-oriented method with cyan bar color and legend of Chicago.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label='Chicago')

# Add the legend.
ax.legend()

# Add a title and axes labels.
plt.title('PyBer Far by Month')
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Invert the y-axis data.
plt.gca().invert_yaxis()
```

```python
# Create a scatter plot with the MATLAB method number 1.
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot with the MATLAB method number 2.
plt.scatter(x_axis, y_axis)
```

```python
# Create a scatter plot with MATLAB that has legend labeled Chicago and red markers. Fare data on the x-axis.
plt.scatter(y_axis, x_axis, c="red", label="Chicago")

# Add the legend.
plt.legend()

# Add a title and label the axes.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add limit to the x-axis.
plt.xlim(0, 45)

# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# Create a Bubble Chart using MATLAB.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Iterate through the y-axis data and increase the data point by 3 to make the bubbles bigger.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
# Use the new y-axis list for the s parameter.
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Do the above, but with list comprehension.
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a scatter plot with object-oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart with object-oriented method.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a bubble chart with object-oriented method. 
# Sky blue markers, marker size 5 times each data point, color 20% transparent, black edge to circle, linewidth 2 points, label legend Boston.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i * 5 for i in y_axis], c= "skyblue", edgecolor= "black", alpha=.8, linewidth=2, label='Boston')

# Add legend.
plt.legend()

# Add title and axis labels.
plt.title('PyBer Fare by Month')
plt.ylabel('Month')
plt.xlabel('Fare($)')

# Add limit to x-axis.
plt.xlim(0, 45)

# Invert y-axis.
plt.gca().invert_yaxis()
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Increase size, explode July and add percentages
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen",
         "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```

```python
# Create a pie chart using the object-oriented method.
fig, ax = plt.subplots(figsize=(8,8))
colors = ["paleturquoise", "plum", "palevioletred", "palegoldenrod", 
          "cornflowerblue", "palegreen", "lemonchiffon", "peachpuff",
         "aquamarine", "lightcoral", "lightskyblue", "mediumorchid"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis,
      autopct="%.1f%%",
      explode=explode_values,
      shadow=True,
      counterclock=False,
      startangle=96,
      colors=colors)
plt.show()
```
